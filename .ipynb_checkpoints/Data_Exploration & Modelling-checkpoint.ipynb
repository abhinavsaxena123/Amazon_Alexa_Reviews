## Data Cleaning

```python
lemmatizer = WordNetLemmatizer()
```

```python
# Function to convert NLTK POS tags to WordNet POS tags
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to noun if no specific tag is provided
```

```python
def clean_reviews(review):
    clean_words = []
    
    review = review.lower()                   #Convert Text to Lowercase
    review = re.sub('\[.*?\]', '', review)         #Remove Text in Square Brackets
    review = re.sub('https?://\S+|www\.\S+', '', review)    #Remove URLs
    review = re.sub('<.*?>+', '', review)          #Remove HTML Tags
    review = re.sub(r'[^\w\s]', '', review)        #Remove Punctuation
    review = re.sub('\n', '', review)              #Remove Newline Characters
    review= re.sub("\d+(\.\d+)?"," ",review)   #Remove numbers
    
    #Removing StopWords
    stop_words = set(stopwords.words('english'))
    stop_words.remove('not')  # Keep 'not' to preserve negation context
    
    review_tokens = word_tokenize(review)  # Tokenize the review text into words
    
    # Lemmatize each word based on its POS tag
    for word in review_tokens :
      if word.lower() not in stop_words:
        pos = pos_tag([word]) # get the part of speech of each word 
        tag = get_simple_pos(pos[0][1])  # Convert POS tag to WordNet POS tag
        clean_word=lemmatizer.lemmatize(word.lower(),pos=tag)
        clean_words.append(clean_word)

    return " ".join(clean_words)
```

```python
df['verified_reviews_clean'] = df['verified_reviews'].apply(lambda x: clean_reviews(x))
```

## Model Building

```python
X = df['verified_reviews_clean']
y = df['feedback']
```

#### Splitting data into train and test set-->
#### 80% for training
#### 20% for testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y,test_size = 0.2,stratify=y,random_state=42)

print(f"X train: {X_train.shape}")
print(f"y train: {y_train.shape}")
print(f"X test: {X_test.shape}")
print(f"y test: {y_test.shape}")
```

#### Using Count Vectorizer to create a bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range = (1, 2),stop_words='english')
count_train = cv.fit_transform(X_train)
count_test = cv.transform(X_test)
```

```python
#Saving the Count Vectorizer
pickle.dump(cv, open('Models/countVectorizer.pkl', 'wb'))
```

```python
for feature in cv.get_feature_names_out():
    print(feature)
```

```python
def plot_confusion_matrix(cm,title='Confusion Matrix',cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()

    classes = ['Negative', 'Positive'] # 0 ->Negative, 1->Positive
    plt.xticks(ticks=np.arange(len(classes)), labels=classes)
    plt.yticks(ticks=np.arange(len(classes)), labels=classes)

     # Add text annotations to each cell
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, f'{cm[i, j]}',
                     ha='center', va='center',
                     color='white' if cm[i, j] > cm.max() / 2. else 'black')
            
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
```

### Multinomial Naive Bayes with Count Vectorizer

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
#Train the model
mnb_model = MultinomialNB()
mnb_model.fit(count_train,y_train)

# Make predictions
mnb_pred = mnb_model.predict(count_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, mnb_pred)
print(f"Accuracy: {accuracy:.3f}")

mnb_cm = confusion_matrix(y_test, mnb_pred)
plot_confusion_matrix(mnb_cm)
```

```python
# Initialize model_score as a dictionary
model_score = {}

metric1=round(accuracy_score(y_test, mnb_pred)*100,2)
print ("Accuracy:{}".format(metric1))

metric2=round(f1_score(y_test, mnb_pred ,average='macro')*100,2)
print ("F1 score:{}".format(metric2))

metric3=round(recall_score(y_test, mnb_pred, average='macro')*100,2)
print ("Recall:{}".format(metric3))

metric4=round(precision_score(y_test, mnb_pred, average='macro')*100,2)
print ("Precision:{}".format(metric4))

metric5=round(roc_auc_score(y_test, mnb_pred, average='macro')*100,2)
print ("roc_auc_score:{}".format(metric5))

model_score["MNB with Count Vectorizer"] = [metric1, metric2, metric3, metric4, metric5]

print(classification_report(y_test, mnb_pred))
```

#### Class 0 has low recall, indicating that the model misses a lot of class 0(Negative Reviews) instances
